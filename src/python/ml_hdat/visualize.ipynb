```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 아래 코드를 이쁘게 만들어줘

plt.figure(figsize=(5,3)) # 시작폼
plt.plot([2,1,3,1,2])
plt.plot([5,3,1,2,4])
plt.show() # 마침표
```

```python
plt.figure(figsize=(6, 4))

# 첫 번째 선 그래프
plt.plot([2, 1, 3, 1, 2], label='Series 1', color='b', marker='o')

# 두 번째 선 그래프
plt.plot([5, 3, 1, 2, 4], label='Series 2', color='r', marker='x')

# 제목 및 레이블 설정
plt.title('Example Line Plot', fontsize=14)
plt.xlabel('X Axis', fontsize=12)
plt.ylabel('Y Axis', fontsize=12)

# 범례 추가
plt.legend()

# 그리드 추가
plt.grid(True)

# 그래프 출력
plt.show()
```

### 연속된 숫자 데이터 시각화
- 데이터가 어디에 몰려있고, 어떻게 퍼져 있는지 (분포)
- sns.histplot

```python
sns.histplot(data=df, x="tip", bins=20, hue="sex", kde=True)
```

```python
sns.histplot(data=df, x="total_bill", bins=20, hue="smoker")
```

### 카테고리 데이터 : countplot
### 연속 데이터 : histplot
### 연속되 숫자 데이터 vs 카테고리 데이터 사이의 관계
- 카테고리 벼로 연속된 숫자 데이터의 분포
- sns.boxplot, sns.violinplot

```python
sns.boxplot(data=df, x="day", y="tip")
```

```python
sns.boxplot(data=df, x="day", y="tip", hue="sex")
```

```python
sns.violinplot(data=df, x="day", y="tip")
```

```python
sns.violinplot(data=df, x="day", y="total_bill", hue="sex")
```

```python
 sns.boxplot(data=df, x="tip", y="day")
```

### 연속된 숫자 데이터 vs 연속된 숫자 데이터 간의 관계
- 2차원 평면 위에 두 데이터가 만나는 지점마다 점을 찍음 : 두 데이터가 어떻게 퍼져있는가
- 산포도

```python
sns.scatterplot(data=df, x="total_bill", y="tip")
```

```python
sns.scatterplot(data=df, x="total_bill", y="tip", hue="smoker")
```

# 머신러닝 맛보기

```python
df = sns.load_dataset("iris")
df.head()
```

```python
sns.pairplot(data=df, hue="species")
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(data=corr, annot=True, cmap="Spectral", fmt=".2f")
```

### 새로운 데이터가 오면 주변에 데이터를 보고 판단
- 새로운 데이터 : 주변 데이터와 거리가 가까운 데이터 (거리 측정)
- 거리 작은 순으로 정렬
- K 개의 데이터를 가지고 와서
