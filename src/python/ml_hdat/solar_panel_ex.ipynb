제공되는 Baseline Code는 본 인증 평가의 데이터 분석 방법 이해를 돕기 위한 참고용이며,

자율적으로 분석/모델링 방법을 활용하여 평가에 참여하실 수 있습니다.

## 1. 데이터 불러오기

```python
import pandas as pd

train = pd.read_csv('./train.csv', encoding='utf-8')
test = pd.read_csv('./test.csv', encoding='utf-8')
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

## 2. 데이터 전처리
예측 목표인 '태양광 설치 수'를 y로 설정하고, 나머지 변수들을 x로 설정합니다.


다만 x를 구성할 때에는 y와 관련 없는 식별자인 ID 컬럼과


모델이 일반적으로 학습할 수 없는 변수들은 제외합니다.


```python
train.dropna(inplace=True)
train_x = train[['가스 난방 연료 사용 비율',	'전기 난방 연료 사용 비율',	'주택 단위 중간 가치',	'주거용 인센티브 수',	'아시아인 비율',	'평균 가구원 수',	'전체 전기 요금',	'위도',	'총 전력 소비량',	'연료유/등유 난방 연료 사용 비율',	'비주거용 주정부 인센티브',	'일일 태양 복사량',	'경도',	'지역 코드',
]]
train_y = train['태양광 설치 수']
```

## 3. 분석 모델 설계
분석을 위한 모델을 준비합니다.

여기서는 가장 간단한 LinearRegression 모델을 사용하여 회귀 모델을 생성합니다.

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
```

## 4. 모델 학습
분석 모델을 학습시킵니다.

```python
lr_model.fit(train_x, train_y)
```

## 5. 예측값 생성
학습한 모델을 사용하여 예측값을 생성합니다.


```python
test=test.fillna(0)
```

```python
# 테스트 데이터는 학습 데이터와의 구성을 일치시킵니다.
test_x = test[['가스 난방 연료 사용 비율',	'전기 난방 연료 사용 비율',	'주택 단위 중간 가치',	'주거용 인센티브 수',	'아시아인 비율',	'평균 가구원 수',	'전체 전기 요금',	'위도',	'총 전력 소비량',	'연료유/등유 난방 연료 사용 비율',	'비주거용 주정부 인센티브',	'일일 태양 복사량',	'경도',	'지역 코드',
]]
```

```python
prediction = lr_model.predict(test_x)
prediction #예측이 진행된 것을 확인하실 수 있습니다.
```

## 6. 제출 파일 생성
submission 파일을 만들어서 제출합니다.

```python
submission = pd.read_csv('sample_submission.csv')
```

```python
#prediction을 submission에 입력합니다.
submission['태양광 설치 수'] = prediction
```

```python
# submission을 csv 파일로 다운 받습니다.
submission.to_csv('baseline_submit.csv', index=False)
```
